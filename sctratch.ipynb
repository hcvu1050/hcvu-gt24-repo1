```python
import matplotlib.pyplot as plt
import csv
```

```python
csv_file_name = 'training_loss.csv'

# Read the CSV file and convert its content into a Python list
with open(csv_file_name, mode='r') as csv_file:
    reader = csv.reader(csv_file)
    training_loss = list(reader)[0] 
training_loss  = [float(x) for x in training_loss]

csv_file_name = 'validation_loss.csv'

# Read the CSV file and convert its content into a Python list
with open(csv_file_name, mode='r') as csv_file:
    reader = csv.reader(csv_file)
    validation_loss= list(reader)[0] 
validation_loss = [float(x) for x in validation_loss]
```

```python
plt.plot(range(1, len(training_loss) + 1), training_loss, label='Training Loss')
plt.plot(range(1, len(validation_loss) + 1), validation_loss, label='Validation Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.show()
```
